**Rule 1 — Subject–Verb Agreement (SVA)**

```python
from src.rules import subject_verb_agreement_rule
import spacy # Import spaCy, the NLP library used for parsing sentences

# Load the small English language model
# This model provides tokenization, POS tagging, and dependency parsing
nlp = spacy.load("en_core_web_sm")

sentence = "He go to market"
# Convert the sentence into a spaCy Doc object
# The Doc contains tokens, POS tags, dependencies, and morphology
doc = nlp(sentence)


print(subject_verb_agreement_rule(doc))
```
